```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
```

```python
df_clean = df_master.dropna()
print(df_clean)
```

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'วิศวฯ', 'Engineering','วิศวกรรม']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name

df_clean['คณะ_มาตรฐาน'] = df_clean['คณะ'].apply(standardize_faculty)
```

```python
print("ข้อมูลต้นฉบับ")
print(df_clean)
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df_clean['อายุ'].mean():.2f}")

# วิธีง่ายๆ - ใช้ IQR (Interquartile Range)
Q1 = df_clean['อายุ'].quantile(0.25)
Q3 = df_clean['อายุ'].quantile(0.75)
IQR = Q3 - Q1

# กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# กรองค่าผิดปกติ
df_no_outliers = df_clean[(df_clean['อายุ'] >= lower_bound) & (df_clean['อายุ'] <= upper_bound)]

print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_no_outliers)
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) - {df_no_outliers['อายุ'].mean():.2f}")
```
